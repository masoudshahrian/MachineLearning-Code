```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python


# تابعی برای بارگذاری و نمایش تصویر
def load_and_display_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # تغییر رنگ از BGR به RGB
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    return img

# بارگذاری تصویر
image_path = '/content/1 (128).png'  # مسیر به تصویر خود را وارد کنید
image = load_and_display_image(image_path)

# تقسیم تصویر به نیمه‌های بالا و پایین
height, width, _ = image.shape
upper_half = image[:height//2, :]  # نیمه بالایی
lower_half = image[height//2:, :]  # نیمه پایینی

# پردازش‌های مورد نظر روی نیمه‌ها (در اینجا فقط به عنوان مثال، می‌توانید تغییرات دلخواه را انجام دهید)
# به عنوان مثال، می‌توانیم نیمه پایینی را به رنگ سیاه و سفید تبدیل کنیم
lower_half_processed = cv2.cvtColor(lower_half, cv2.COLOR_BGR2GRAY)
lower_half_processed = cv2.cvtColor(lower_half_processed, cv2.COLOR_GRAY2BGR)  # تبدیل به رنگی برای ترکیب

# ترکیب نیمه‌ها
combined_image = np.vstack((upper_half, lower_half_processed))

# نمایش تصویر نهایی
plt.imshow(combined_image)
plt.axis('off')
plt.title('Combined Image')
plt.show()

# ذخیره تصویر نهایی
output_path = '/content/combined_image.jpg'
cv2.imwrite(output_path, cv2.cvtColor(combined_image, cv2.COLOR_RGB2BGR))  # تغییر رنگ به BGR برای ذخیره
print(f"Image saved at: {output_path}")
```
